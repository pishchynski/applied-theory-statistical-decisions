```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 3
N = 100
means = []
covs = []
res = []
for i in range(L):
    means.append((i + 1) * 8 * np.random.random_sample(N))
    covs.append(np.eye(N))
    res.append(np.random.multivariate_normal(mean=means[i], cov=covs[i]))
```

```python
for i in range(L):
    plt.plot(res[i], 'o')
plt.show()
```
